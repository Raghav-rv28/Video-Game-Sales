# Video Game Data Insights ``(EDA)``

```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import os
filepath = "./Datasets/vgsales.csv"
df = pd.read_csv(filepath,index_col='Rank')
df.head(5)
```

```python
df.info()
```

```python
df.shape
```

```python
missing = pd.DataFrame((df.isnull().sum())*100/df.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot('index',0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

```python
df.dropna(how="any",inplace=True)
```

```python
missing = pd.DataFrame((df.isnull().sum())*100/df.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot('index',0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

```python
df.loc[:,['Year']] = df.Year.astype(int)
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(16,6))
plt.xlabel("Sales")
plt.ylabel("Sales in Millions")
plt.title("Top Selling Games by Year of Release")
sns.lineplot(df.Year,df.Global_Sales)
```

# Video Game Sales Data Insights
- Top Selling Games by Genre
- Top Selling Games by Year
- Top Selling Games by Platform
- Top Selling Games by Publisher

```python
TopSellingGamesByGenre = df.sort_values(by=["Global_Sales"]).groupby(by=["Genre"])['Name','Global_Sales'].max()
plt.figure(figsize=(20,6))
plt.legend("Top Selling Games By Genre")
sns.barplot(TopSellingGamesByGenre['Global_Sales'],TopSellingGamesByGenre['Name'])
plt.show()
TopSellingGamesByPublisher= df.sort_values(by=["Global_Sales"]).groupby(by=["Publisher"])['Name','Global_Sales'].max()
TopSellingGamesByPublisher.sort_values(by=['Global_Sales'],ascending=False)
plt.figure(figsize=(20,6))
plt.legend("Top 20 Selling Games By Publisher")
sns.barplot(TopSellingGamesByPublisher.Global_Sales.iloc[:25],TopSellingGamesByPublisher.Name.iloc[:25])
plt.show()
TopSellingGamesByPlatform= df.sort_values(by=["Global_Sales"]).groupby(by=["Platform"])['Name','Global_Sales'].max()
TopSellingGamesByPlatform.sort_values(by=['Global_Sales'],ascending=False)
plt.figure(figsize=(20,6))
plt.legend("Top Selling Games By Platform")
sns.barplot(TopSellingGamesByPlatform.Global_Sales.iloc[:25],TopSellingGamesByPlatform.Name.iloc[:25])
plt.show()
```

## More Data Insights
- Top 10 Games by Global Sales
- Top 10 Games by NA Sales
- Top 10 Games by JP Sales
- Top 10 Games by EU Sales

```python
Topten_byGlobalSales = df.loc[:,['Global_Sales','Name']].sort_values(by=["Global_Sales"],ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.legend("Top 10 Games By Global Sales")
sns.barplot(Topten_byGlobalSales['Global_Sales'],Topten_byGlobalSales['Name'])
plt.show()
Topten_byEUSales = df.loc[:,['EU_Sales','Name']].sort_values(by=["EU_Sales"],ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.legend("Top 10 Games By EU Sales")
sns.barplot(Topten_byEUSales['EU_Sales'],Topten_byEUSales['Name'])
plt.show()
Topten_byNASales = df.loc[:,['NA_Sales','Name']].sort_values(by=["NA_Sales"],ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.legend("Top 10 Games By NA Sales")
sns.barplot(Topten_byNASales['NA_Sales'],Topten_byNASales['Name'])
plt.show()
Topten_byJPSales = df.loc[:,['JP_Sales','Name']].sort_values(by=["JP_Sales"],ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.legend("Top 10 Games By JP Sales")
sns.barplot(Topten_byJPSales['JP_Sales'],Topten_byJPSales['Name'])
plt.show()
Topten_byOtherSales = df.loc[:,['Other_Sales','Name']].sort_values(by=["Other_Sales"],ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.legend("Top 10 Games By Other Sales")
sns.barplot(Topten_byOtherSales['Other_Sales'],Topten_byOtherSales['Name'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("Sales Column")
plt.title("Top 10 Game Sales Percent by Regions")
sns.heatmap(data=df.loc[:,['JP_Sales','NA_Sales','EU_Sales','Other_Sales',"Global_Sales"]].sort_values(by=['Global_Sales'],ascending=False)[:10], annot=True)
```
